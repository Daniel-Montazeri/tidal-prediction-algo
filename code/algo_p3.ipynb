```python
#CALCULATING ROOT MEAN SQUARE ERROR BETWEEN ACTUAL DATA AND DIGITAL PREDICTION
def calculate_rmse(actualData, digitalPrediction):

    # Calculate the residuals and square them
    squared_diff = (actualData - digitalPrediction) ** 2

    # Calculate average of residuals
    mean_squared_diff = np.mean(squared_diff)

    # Calculate the square root of previous result
    rmse = np.sqrt(mean_squared_diff)

    return rmse

#Define actual data and digital prediction
actualData =  june_data['SLEV(metres)']
digitalPrediction = new_wave

rmse_value = calculate_rmse(actualData, digitalPrediction)
print("The Root Mean Square Error between the Actual Data and Digital Prediction is:", rmse_value)
```
